```python
# Preprocessing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_excel (r'Superstore.xls')
```

## Task 10:
### Finding out the majority and minority customer cities on basis of

#### a.) Number of Customers

```python
cities_data_cust = data.groupby(['City'])['Customer Name'].count()
cities_data_cust = pd.DataFrame(cities_data_cust)
cities_data_cust = cities_data_cust.reset_index()
cities_data_cust = cities_data_cust.rename(columns = {"Customer Name": "No. of Customers"})
print("List of cities with no. of customers")
cities_data_cust
```

```python
max_cust_city_index = cities_data_cust['No. of Customers'].idxmax()
max_cust_city = cities_data_cust.iloc[max_cust_city_index]
print("City with the highest number of customers.")
max_cust_city
```

```python
min_cust_city_index = cities_data_cust['No. of Customers'].idxmin()
min_cust_city = cities_data_cust.iloc[min_cust_city_index]
print("City with the lowest number of customers.")
min_cust_city
```

#### b.) Sales Value

```python
cities_data_sales = data.groupby(['City'])['Sales'].sum()
cities_data_sales = pd.DataFrame(cities_data_sales)
cities_data_sales = cities_data_sales.reset_index()
print("List of cities with their total sales")
cities_data_sales.head()
```

```python
max_sales_city_index = cities_data_sales['Sales'].idxmax()
max_sales_city = cities_data_sales.iloc[max_sales_city_index]
print("City with the highest sales.")
max_sales_city
```

```python
min_sales_city_index = cities_data_sales['Sales'].idxmin()
min_sales_city = cities_data_sales.iloc[min_sales_city_index]
print("City with the lowest sales.")
min_sales_city
```

#### c.) Number of quantity sold

```python
cities_data_quant = data.groupby(['City'])['Quantity'].sum()
cities_data_quant = pd.DataFrame(cities_data_quant)
cities_data_quant = cities_data_quant.reset_index()
print("List of cities with their total quantity")
cities_data_quant
```

```python
max_quant_city_index = cities_data_quant['Quantity'].idxmax()
max_quant_city = cities_data_quant.iloc[max_quant_city_index]
print("City with highest quantities sold")
max_quant_city
```

```python
min_quant_city_index = cities_data_quant['Quantity'].idxmin()
min_quant_city = cities_data_quant.iloc[min_quant_city_index]
print("City with lowest quantities sold")
min_quant_city
```
