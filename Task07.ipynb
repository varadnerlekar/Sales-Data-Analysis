```python
# Preprocessing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_excel (r'Superstore.xls')
data
```

## Task 7:
### Finding out the most and least sold product ID

#### a.) Create a new dataset including the product IDs and total quantities sold for that ID

```python
prod_id = data.groupby(['Product ID','Category','Sub-Category'])['Quantity'].sum()
prod_id = pd.DataFrame(prod_id)
prod_id.sort_values(by=['Product ID'], inplace=True)
prod_id = prod_id.reset_index()
print("List of Product IDs and their quantities.")
prod_id
```

#### b.) Find out the most sold product from the created dataset

```python
most_sold_quant = prod_id['Quantity'].max()
most_sold_index = prod_id['Quantity'].idxmax()
most_sold = prod_id.iloc[most_sold_index]
print("Information of Most Sold Product")
most_sold
```

```python
least_sold_quant = prod_id['Quantity'].min()
least_sold_index = prod_id['Quantity'].idxmin()
least_sold = prod_id.iloc[least_sold_index]
print("Information of Least Sold Product")
least_sold
```
