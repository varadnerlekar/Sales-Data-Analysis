```python
# Preprocessing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_excel (r'Superstore.xls')
```

## Task 8:
### Finding out the customer who bought most and least from us in terms of quantity

#### a.) Create a dataset containing name and quantities bought

```python
customer = data.groupby(['Customer ID','Customer Name'])['Quantity'].sum()
customer = pd.DataFrame(customer)
customer.sort_values(by=['Customer ID'], inplace=True)
customer = customer.reset_index()
print("Customer List")
customer
```

#### b.) Find out the customer name and quantity, who bought maximum in quantity

```python
max_purchase_customer_quant = customer['Quantity'].max()
max_purchase_customer_index = customer['Quantity'].idxmax()
max_purchase_customer = customer.iloc[max_purchase_customer_index]
print("Information of the customer who has bought the most in termns of quantity.")
max_purchase_customer
```

```python
min_purchase_customer_quant = customer['Quantity'].min()
min_purchase_customer_index = customer['Quantity'].idxmin()
min_purchase_customer = customer.iloc[min_purchase_customer_index]
print("Information of the customer who has bought the least in termns of quantity.")
min_purchase_customer
```
