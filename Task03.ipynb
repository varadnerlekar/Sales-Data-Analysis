```python
# Preprocessing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_excel (r'Superstore.xls')
```

## Task 3:
### Find out the per unit price from the data

#### a.) Convert data time field into pandas time object

```python
data["Order Date"] = pd.to_datetime(data["Order Date"])
data["Ship Date"] = pd.to_datetime(data["Ship Date"])
data.info()
```

Converted Order Date and Ship Date into Pandas DateTime Object

#### b.) Find out the price per unit from the data and create a column for it in the same data

```python
data['Price Per Unit'] = data["Sales"] / data["Quantity"]
print("This is the column indicating the Price per unit data.")
pd.DataFrame(data['Price Per Unit']).head()
```
