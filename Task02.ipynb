### d.) Print the attributes like count, mean, max, min, standard deviation etc

```python
Count = data.count()
Count = pd.DataFrame(Count)
Mean = data.mean()
Mean = pd.DataFrame(Mean)
Max = data.max()
Max = pd.DataFrame(Max)
Min = data.min()
Min = pd.DataFrame(Min)
Standard_Deviation = data.std()
Standard_Deviation = pd.DataFrame(Standard_Deviation)

attributes = [Count,Mean,Max,Min,Standard_Deviation]

for i in range(len(attributes)):
    Count = Count.merge(attributes[i], left_index=True, right_index=True, suffixes=(str(i),str(i+1)))
        
Count = Count.drop(columns=["00"])         
Count = Count.rename(columns={"01":"Count", "02":"Mean", "03":"Max", "04":"Min", "05":"Standard Deviation"})
attribute_data = Count
attribute_data = pd.DataFrame(attribute_data)
attribute_data
```
